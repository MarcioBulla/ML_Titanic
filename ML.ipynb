#### Encoder
Criaremos um encoder para os titulos e refatoremos para uma aplicação mais pratica

```python
title_encoder = LabelEncoder().fit(data.Title.unique().tolist())

print(title_encoder.classes_)
data["Title_code"] = title_encoder.transform(data.Title)
print(data[["Title_code", "Title"]].nunique())

# salvar Encoder em um arquivo
jb.dump(title_encoder, os.path.join(local, "title_econder.pkl"))
```

#### Refatoração

```python
def ref_title(df_origin, df):
    df_origin["Title"] = df_origin.Name.str.extract(r"([A-Za-z]+)\.")
    
    # funções para diminuir a quantidade valores unicos
    Mlle_Miss_Mme = lambda x: "Miss" if x.lower() in "Mlle_Miss_Mme_Countess_Lady_Ms".lower().split("_") else x
    Other = lambda x: "Othet" if x.lower() not in "Mr, Mrs, Miss".lower().split(", ") else x

    df_origin.Title = df_origin.Title.apply(Mlle_Miss_Mme)
    df_origin.Title = df_origin.Title.apply(Other)
    
    # encoder applie
    df_origin.loc[:, "Title"] = title_encoder.transform(df_origin.Title)
    
    # applied on train and test dataframe 
    df["Title"] = df_origin["Title"]
```

```python
def ref_age(df_origin, df):
    for sex, pclass in [(sex, pclass) for sex in df_origin.Sex.unique() for pclass in df_origin.Pclass.unique()]:
        condition = (df_origin.Sex == sex) & (df_origin.Pclass == pclass)
        age_mean = df_origin.loc[condition, "Age"].mean()
        df_origin.loc[condition & df_origin.Age.isnull(), "Age"] = age_mean
    
    df["Age"] = df_origin.Age
```

## Preparativos

```python
seed = 42

# Divisões no Cros validate
CV = 7

# DataFrame para salvar o desempenho dos modelos
models_default = pd.DataFrame()
```

### Criando modelos 

```python
lr = LogisticRegression(random_state=seed)
dt = DecisionTreeClassifier(random_state=seed)
rf = RandomForestClassifier(random_state=seed)
svc = SVC(random_state=seed)
knn = KNeighborsClassifier()

models = [lr, dt, rf, svc, knn]

xgb = XGBClassifier(random_state=seed)
```

### Funções para testar

Funções para 

```python
def testing_models(models, df_features, df_target, CV):
    result_models = pd.DataFrame()
    
    for model in models:
        # Cross Validation
        result_models[type(model).__name__] = cross_val_score(model, df_features, df_target, cv=CV)
    
    return result_models

def testing_XGB(xgb, df_features, df_target, CV):
    global seed
    kf = KFold(n_splits=CV, shuffle=True, random_state=seed)
    score_xgb = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X_norm.iloc[train_index], X_norm.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        xgb.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=0)
        score_xgb.append(xgb.score(X_test, y_test))

    return score_xgb
```

Função para comparar modelos

```python
def comparing_models(df):
    return pd.concat([df.mean(), df.var()], axis=1).rename(columns={0:"Mean", 1:"std"}).sort_values(by=["Mean", "std"], ascending=[False, False]).style.background_gradient(subset=["Mean", "std"])
```

## Stacking dos melhores

### Preparativos

```python
def stacking(df_feature, df_target, models, META, CV):
    global seed
    sclf = StackingCVClassifier(classifiers=models, meta_classifier=META, random_state=seed, n_jobs=-1)
    return (sclf, cross_val_score(sclf, df_feature, df_target, cv=CV, scoring='accuracy', n_jobs=-1))

def best_META(df_feature, df_target, models, CV):
    global seed
    best_Meta = {"Meta": None, "Mean" : 0}
    for META in models:
        mean = stacking(df_feature, df_target, models, META, CV)[1].mean()
        if  mean > best_Meta["Mean"]:
            best_Meta["Mean"] = mean
            best_Meta["Meta"] = META
    return best_Meta["Meta"]

def stacking_best(df_feature, df_target, models, CV):
    global seed
    best_Meta = best_META(df_feature, df_target, models, CV)
    print(f"Best Meta: {type(best_Meta).__name__}")
    return stacking(df_feature, df_target, models, best_Meta, CV)
```
